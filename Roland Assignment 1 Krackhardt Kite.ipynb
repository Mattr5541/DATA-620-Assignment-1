## Krackhardt Kite

```python
import networkx as nx
import matplotlib.pyplot as plt

kite = nx.krackhardt_kite_graph()

labels = {0:'Billy-Bob', 1:'Pucci', 2:'Argyle', 3:'Joe-Bob', 4:'Timothy', 5:'Liam', 6:'Jolyne', 7:'Matthias', 8:'Alice', 9:'Gail'}

plt.figure(figsize = (10,10))
pos = nx.spring_layout(kite)
nx.draw(kite, pos, labels = labels, node_size = 1000)
plt.title('Network')
plt.show()
```
